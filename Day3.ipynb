Create a numpy array starting from 2 till 50 with a stepsize of 3.

### Question-2

Accept two lists of 5 elements each from the user.
Convert them to numpy arrays. Concatenate these arrays and print it. Also sort these arrays and print it.

```python
import numpy as np
list1 = []
list2 = []

print("Enter List1:")
for i in range(5):
    num = int(input())
    list1.append(num)

print("Enter List2:")
for i in range(5):
    num = int(input())
    list2.append(num)

array1 = np.array(list1)
array2 = np.array(list2)

concat_array = np.concatenate((array1, array2))
print(concat_array)
print(np.sort(concat_array))
```

### Question-4

How to convert a 1D array into a 2D array? Demonstrate with the help of a code snippet
Hint: np.newaxis, np.expand_dims

```python
import numpy as np
array = np.arange(9)
newarray = array.reshape(3,3)
print(newarray)
```

In above code, reshape contain 2 parameters
 - first parameter: dimensions like row size
 - second parameter: number of elements like column size
dimension and element number need to match with array size

if array size is 8 and tried to (3,3) then it will not be allowed (it requires 3*3 = 9 array size)

### Question-5

Consider two square numpy arrays. Stack them vertically and horizontally.
Hint: Use vstack(), hstack()

```python
import numpy as np
array1 = np.square([[1,2,3],[4,5,6]])
array2 = np.square([[7,8,9],[10,11,12]])
print(np.vstack((array1, array2)))
print(np.hstack((array1, array2)))
```

### Question-6

How to get unique items and counts of unique items?

```python
import numpy as np
array = np.array([1,2,3,2,4,8,3,1])
unique, count = np.unique(array, return_counts = True)
print(unique)
print(count)
```
