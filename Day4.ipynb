### Question-2

How to create a series from a numpy array?

```python
import numpy as np
import pandas as pd

numpy_array = np.array([1,2,3,4,5])
pandas_series = pd.Series(numpy_array)
print(pandas_series)
```

### Question-3

How to convert the index of a series into a column of a dataframe?

```python
import numpy as np
import pandas as pd

numpy_array = np.array([5,6,7,25,45,63,74,12])
series = pd.Series(numpy_array)
pd.DataFrame(series, columns=['NewColumn'])
```
