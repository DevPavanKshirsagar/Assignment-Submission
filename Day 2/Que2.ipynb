# List comprehension

List comprehension is used to write a code in single line.
- it is used to create a list based on existing lists.
- it is more faster than normal functions.
- but if we write very long list comprehension in single line then code seems to be little bit harder to read.

Simple example: multiplication of first 10 num with 2
- Normal code:

```python
list = []
for i in range(10):
    list.append(i*2)
print(list)
```

- List comprehension code:

```python
list = [i * 2 for i in range(10)]
print(list)
```

##### Syntax of list comprehension:

[expression for item in list]

references from above example:
  - expression : i * 2
  - item : i
  - list : range(10) - 0~(10-1) num

It is possible to add printable text into the expression also

Example: even odd number

```python
data = [str(i) + ' is Even' if i % 2 == 0 else str(i) + " is odd" for i in range(10)]
print(data)
```

#### Nested loops are also is supported in list comprehensions
